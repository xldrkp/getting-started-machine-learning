# matplotlib

matplotlib ist ein Standard, wenn es um die Visualisierung von Daten mit Python geht. In diesem Abschnitt Lernen Sie die Grundlagen kennen, wobei Sie sich [am offiziellen Tutorial](https://matplotlib.org/stable/tutorials/introductory/pyplot.html) der Bibliothek orientieren.

## Ein erstes Beispiel

```python
import matplotlib.pyplot as plt
plt.plot([0, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

## Formatieren der Visualisierung

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

## Ein Beispiel mit *numpy*

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
